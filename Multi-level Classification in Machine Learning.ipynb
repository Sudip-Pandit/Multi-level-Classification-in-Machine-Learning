### Print the features names

```python
x = iris.data
y = iris.target
```

```python
df = pd.DataFrame(x, columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
df['target'] = y
df.head()
```

```python
import seaborn as sns
sns.pairplot(df)
```

### Split the dataset into train and test splits

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y)
```

### Model building

```python
clf = DecisionTreeClassifier(criterion = 'gini', random_state=1)
```

```python
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

### Check the accuracy score

```python
from sklearn import metrics
print('Accuracy:', metrics.accuracy_score(y_test, y_pred))
```

### Print the confusion matrix

```python
print('Confusion Matrix')
mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(mat, square=True, annot=True, fmt = 'd', cbar = False, xticklabels = iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
mat
```

### Classification Report

```python
print(metrics.classification_report(y_test, y_pred))
```
